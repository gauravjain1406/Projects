```python
# We calculate the mean of the stock price at the end of the each day
reliance['avgstockpriceforeachday']=(reliance['Close']+reliance['Open'])/2
```

```python
#Calculate the mean stock price

reliance['avgstockpriceforeachday'].mean()
```

```python
#lets calculate the diff between mean price and the stock price for each day.
reliance['diffpriceandmeanprice']=1965.801-reliance['avgstockpriceforeachday']
reliance
```

```python
reliance=reliance.drop(['mean'],axis=1)
```

```python
reliance
```

```python
reliance['square']=reliance['diffpriceandmeanprice']**2
reliance
```

```python
reliance['square'].sum()
```

```python
#variance
Variance=reliance['square'].sum()/1236
```

```python
Variance
```

```python
#StandardDeviation
StandardDeviation=Variance**0.5
StandardDeviation
```

```python
#It means the stock price of reliance usually deviates from the mean price is 483.999 in 5 years.
```

```python
percentvotality=(StandardDeviation/1965.801)*100
percentvotality
```

```python
#24.62% votality rate is high which means reliance stock market is risky if we consider 5 years time period.
```
